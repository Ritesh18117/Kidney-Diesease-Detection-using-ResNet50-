```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import numpy as np
import os
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://chatgpt.com/c/68a75c5e-23f0-832e-829b-0c7f9108293b
```

```python
# -----------------------
# Paths & params
# -----------------------
train_dir = "dataset/train"
val_dir   = "dataset/val"
test_dir  = "dataset/test"

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
# -----------------------
# Count images in each folder (before balancing)
# -----------------------
def count_images(directory):
    class_counts = {}
    for class_name in os.listdir(directory):
        class_path = os.path.join(directory, class_name)
        if os.path.isdir(class_path):
            class_counts[class_name] = len(os.listdir(class_path))
    return class_counts

train_counts = count_images(train_dir)
test_counts  = count_images(test_dir)
val_counts   = count_images(val_dir)

print("Train Counts:", train_counts)
print("Test Counts:", test_counts)
print("Validation Counts:", val_counts)
```

```python
# Plot class distribution before balancing
def plot_class_distribution(counts_dict, title):
    plt.figure(figsize=(6,4))
    sns.barplot(x=list(counts_dict.keys()), y=list(counts_dict.values()))
    plt.title(title)
    plt.ylabel("Number of Images")
    plt.xlabel("Classes")
    plt.show()

plot_class_distribution(train_counts, "Train Data Distribution (Before Balancing)")
plot_class_distribution(test_counts, "Test Data Distribution")
plot_class_distribution(val_counts, "Validation Data Distribution")
```

```python
# -----------------------
# Generators (use proper preprocessing for ResNet!)
# -----------------------
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode="nearest"
)

test_val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_data = train_datagen.flow_from_directory(
    train_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=True
)

val_data = test_val_datagen.flow_from_directory(
    val_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=False
)

test_data = test_val_datagen.flow_from_directory(
    test_dir, target_size=IMG_SIZE, batch_size=BATCH_SIZE,
    class_mode="categorical", shuffle=False
)
```

```python
# -----------------------
# Handle Class Imbalance using Class Weights
# -----------------------
# Compute class weights from training data
counter = Counter(train_data.classes)
max_val = float(max(counter.values()))
class_weights = {cls: max_val / num_images for cls, num_images in counter.items()}

print("Class Weights:", class_weights)

# Plot class distribution after balancing (using oversampling logic)
balanced_counts = {cls: int(max_val) for cls in counter.keys()}
class_labels = list(train_data.class_indices.keys())
balanced_counts_named = {class_labels[k]: v for k,v in balanced_counts.items()}
plot_class_distribution(balanced_counts_named, "Train Data Distribution (After Balancing - Weighted)")
```

```python
# -----------------------
# Show Sample Images from Each Class
# -----------------------
def show_sample_images(generator, class_labels):
    plt.figure(figsize=(12,8))
    for i, label in enumerate(class_labels):
        # Get one image from each class folder
        class_dir = os.path.join(train_dir, label)
        img_name = os.listdir(class_dir)[0]  # first image
        img_path = os.path.join(class_dir, img_name)
        img = tf.keras.utils.load_img(img_path, target_size=IMG_SIZE)
        plt.subplot(1,4,i+1)
        plt.imshow(img)
        plt.title(label)
        plt.axis("off")
    plt.suptitle("Sample Images from Each Class")
    plt.show()

class_labels = list(train_data.class_indices.keys())
show_sample_images(train_data, class_labels)
```

```python
# -----------------------
# Class weights (keep balancing)
# -----------------------
from collections import Counter
counter = Counter(train_data.classes)
max_val = float(max(counter.values()))
class_weights = {cls: max_val / num_images for cls, num_images in counter.items()}
print("Class weights:", class_weights)
```

```python
# -----------------------
# Model
# -----------------------
base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(224,224,3))
base_model.trainable = False  # Stage 1: freeze

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, activation="relu"),
    layers.Dropout(0.4),
    layers.Dense(4, activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
# -----------------------
# Callbacks
# -----------------------
ckpt = ModelCheckpoint("resnet50_best.h5", monitor="val_accuracy", save_best_only=True, verbose=1)
es   = EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True, verbose=1)
rlr  = ReduceLROnPlateau(monitor="val_loss", factor=0.2, patience=3, min_lr=1e-6, verbose=1)
```

```python
# -----------------------
# Stage 1 training (head only)
# -----------------------
history1 = model.fit(
    train_data, epochs=8, validation_data=val_data,
    class_weight=class_weights, callbacks=[ckpt, rlr, es]
)
```

```python
# -----------------------
# Stage 2: Fine-tune top of ResNet
# -----------------------
# Unfreeze last ~50 layers of backbone
base_model.trainable = True
for layer in base_model.layers[:-50]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5),  # much lower LR
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history2 = model.fit(
    train_data, epochs=15, validation_data=val_data,
    class_weight=class_weights, callbacks=[ckpt, rlr, es]
)
```

```python
# -----------------------
# Curves
# -----------------------
def plot_hist(h, title):
    plt.figure(figsize=(12,5))
    # Loss
    plt.subplot(1,2,1)
    plt.plot(h.history["loss"], label="Train Loss")
    plt.plot(h.history["val_loss"], label="Val Loss")
    plt.title(title + " - Loss"); plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()
    # Acc
    plt.subplot(1,2,2)
    plt.plot(h.history["accuracy"], label="Train Acc")
    plt.plot(h.history["val_accuracy"], label="Val Acc")
    plt.title(title + " - Accuracy"); plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
    plt.show()

plot_hist(history1, "Stage 1")
plot_hist(history2, "Stage 2 (Fine-tune)")
```

```python
# -----------------------
# Eval: Confusion Matrix & F1 on VAL (or TEST when youre ready)
# -----------------------
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
val_data.reset()
y_true = val_data.classes
class_labels = list(val_data.class_indices.keys())
y_pred_probs = model.predict(val_data)
y_pred = np.argmax(y_pred_probs, axis=1)

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_labels)
plt.figure(figsize=(6,6))
disp.plot(values_format="d")
plt.title("Confusion Matrix (Validation)")
plt.show()
```

```python

print("Classification Report (Validation):")
print(classification_report(y_true, y_pred, target_names=class_labels))
```

```python
# -----------------------
# Prediction function
# -----------------------
def predict_image(img_path, dataset_path):
    # Create a temporary generator only to fetch class labels
    datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    temp_gen = datagen.flow_from_directory(
        dataset_path,
        target_size=(224,224),
        batch_size=32,
        class_mode="categorical"
    )
    class_labels = list(temp_gen.class_indices.keys())

    # Load + preprocess image
    img = tf.keras.utils.load_img(img_path, target_size=(224, 224))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    # Predict
    preds = model.predict(img_array)
    predicted_index = np.argmax(preds)
    predicted_class = class_labels[predicted_index]
    confidence = np.max(preds) * 100

    # Show result
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Prediction: {predicted_class} ({confidence:.2f}%)")
    plt.show()

    return predicted_class, confidence
```

```python
# predict_image("dataset/val/Normal/Normal- (60).jpg")
# -----------------------
# Example usage
# -----------------------
# Replace with your image path
result, conf = predict_image("Cyst- (1318).jpg")
print("Predicted:", result, "Confidence:", conf)
```
